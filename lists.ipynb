Create a dictionary to store product information (name, price, quantity). Allow users to add, update, and remove items from the inventory.

```python
store = []

def add():
    name = input("Enter product name: ")
    price = input("Enter product price (e.g., $5): ")
    quantity = input("Enter product quantity: ")
    store.append({'name': name, 'price': price, 'quantity': quantity})
    print(f"{name} added\n")

def update():
    name = input("Update Product: ")
    for item in store:
        if item['name'] == name:
            new_price = input("Enter new price: ")
            new_quantity = input("Enter new quantity: ")
            if new_price:
                item['price'] = new_price
            if new_quantity:
                item['quantity'] = new_quantity
            print(f"{name} updated.\n")
            return
    print(f"{name} not found.\n")

def delete():
    name = input("Enter product name to remove: ")
    for item in store:
        if item['name'] == name:
            store.remove(item)
            print(f"{name} removed.\n")
            return
    print(f"{name} not found.\n")

def inventory():
    if not store:
        print("Inventory is empty.\n")
    else:
        print("\ninventory")
        for item in store:
            print(f"Name: {item['name']}, Price: {item['price']}, Quantity: {item['quantity']}")
        print()

def done():
    while True:
        print("1. add product")
        print("2. update product")
        print("3. remove product")
        print("4. view inventory")
        print('5. exit')
        choice = input("Choose an option (1-5): ")

        if choice == '1':
            add()
        elif choice == '2':
            update()
        elif choice == '3':
            delete()
        elif choice == '4':
            inventory()
        elif choice == '5':
            print("Ok bye bye")
            break

done()
```
